## Bonus Question!
#### There are too many decimal places for mean and stddev in the describe() dataframe. Format the numbers to just show up to two decimal places. Pay careful attention to the datatypes that .describe() returns, we didn't cover how to do this exact formatting, but we covered something very similar. [Check this link for a hint](http://spark.apache.org/docs/latest/api/python/pyspark.sql.html#pyspark.sql.Column.cast)

If you get stuck on this, don't worry, just view the solutions.

```pyspark
walmart.withColumn("Open",round(walmart.Open.cast(FloatType()),2)).show(truncate=True)
walmart.withColumn("High",round(walmart.High.cast(FloatType()),2)).show(truncate=True)
walmart.withColumn("Low",round(walmart.Low.cast(FloatType()),2)).show(truncate=True)
walmart.withColumn("Close",round(walmart.Close.cast(FloatType()),2)).show(truncate=True)
walmart.withColumn("Volume",round(walmart.Volume.cast(FloatType()),2)).show(truncate=True)

```

```pyspark
walmart.select(format_number(walmart["Open"].cast("float"), 2).alias("Open"),
format_number(walmart["High"].cast("float"), 2).alias("High"),
format_number(walmart["Low"].cast("float"), 2).alias("Low"),
format_number(walmart["Close"].cast("float"), 2).alias("Close"),
walmart["Volume"].cast("int").alias("Volume")
).show()
```

```pyspark
walmart.printSchema()
```

#### Create a new dataframe with a column called HV Ratio that is the ratio of the High Price versus volume of stock traded for a day.

```pyspark
HVAC_Ratio = walmart.withColumn("HV_Ratio", walmart['High']/walmart['Volume']).show()
```

#### What day had the Peak High in Price?

```pyspark
#add two [0]'s instead of just one to provide the result only
x= walmart.agg({'High':'max'}).collect()[0][0]
walmart.filter(walmart.High == x).select("Date").show()
```

#### What is the mean of the Close column?

```pyspark
walmart.agg({"Close": "avg"}).collect()[0][0]
```

#### What is the max and min of the Volume column?

```pyspark
walmart.agg({'Volume':'max'}).collect()[0][0]
```

```pyspark
walmart.agg({'Volume':'min'}).collect()[0][0]
```

#### How many days was the Close lower than 60 dollars?

```pyspark
walmart.filter(walmart.Close < 60).select("Date").count()
```

#### What percentage of the time was the High greater than 80 dollars ?
#### In other words, (Number of Days High>80)/(Total Days in the dataset)

```pyspark
num_80 = walmart.filter(walmart.High>80).select("Date").count()
count_days = walmart.count()
print(float(num_80)/float(count_days)*100)
```

#### What is the Pearson correlation between High and Volume?
#### [Hint](http://spark.apache.org/docs/latest/api/python/pyspark.sql.html#pyspark.sql.DataFrameStatFunctions.corr)

```pyspark
walmart.select(corr("High", "Volume")).show()
```

#### What is the max High per year?

```pyspark
import datetime
from pyspark.sql.functions import year, month, dayofmonth
walmart.groupBy(year("Date")).agg({'High':'max'}).alias('High_Per_Year').show()
```

#### What is the average Close for each Calendar Month?
#### In other words, across all the years, what is the average Close price for Jan,Feb, Mar, etc... Your result will have a value for each of these months. 

```pyspark
walmart.groupBy(month('Date').alias('month')).agg({'Close':'avg'}).sort('month').show()
```
